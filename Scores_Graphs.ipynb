```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data from the tables
companies = ['Adani Green\nEnergy Ltd.', 'CESC', 'ITC', 'TATA POWER', 'NHPC', 'NTPC']
fiscal_years = ['21-22', '22-23', '23-24']

# Total Scores data
total_scores = {
    'Adani Green\nEnergy Ltd.': [4.955, 5.2125, 5.88],
    'CESC': [4.99, 4.4975, 4.3975],
    'ITC': [5.28, 5.1, 5.505],
    'TATA POWER': [6.2, 6.61, 6.7975],
    'NHPC': [4.46, 5.5775, 3.3725],
    'NTPC': [5.7, 5.5125, 5.88]
}

# Parameters Available data (actual numbers out of 26)
parameters_available = {
    'Adani Green\nEnergy Ltd.': [20, 22, 22],
    'CESC': [22, 22, 23],
    'ITC': [12, 13, 14],
    'TATA POWER': [21, 24, 24],
    'NHPC': [18, 19, 13],
    'NTPC': [17, 17, 16]
}

# Set up the plotting parameters
x = np.arange(len(companies))
width = 0.25

# Create the first chart - Total Scores
fig1, ax1 = plt.subplots(figsize=(14, 8))

bars1 = ax1.bar(x - width, [total_scores[company][0] for company in companies], width, label='FY 21-22', color='#2E86AB', alpha=0.8)
bars2 = ax1.bar(x, [total_scores[company][1] for company in companies], width, label='FY 22-23', color='#A23B72', alpha=0.8)
bars3 = ax1.bar(x + width, [total_scores[company][2] for company in companies], width, label='FY 23-24', color='#F18F01', alpha=0.8)

ax1.set_xlabel('Companies', fontsize=12, fontweight='bold')
ax1.set_ylabel('Total ESG Scores', fontsize=12, fontweight='bold')
ax1.set_title('ESG Total Scores by Company Across Fiscal Years', fontsize=14, fontweight='bold', pad=20)
ax1.set_xticks(x)
ax1.set_xticklabels(companies, rotation=45, ha='right')
ax1.legend(loc='upper left', fontsize=10)
ax1.grid(axis='y', alpha=0.3)
ax1.set_ylim(0, 10)

# Add value labels on bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + 0.05,
                f'{height:.2f}', ha='center', va='bottom', fontsize=8)

add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)

plt.tight_layout()
plt.show()

# Create the second chart - Parameters Available
fig2, ax2 = plt.subplots(figsize=(14, 8))

bars1_param = ax2.bar(x - width, [parameters_available[company][0] for company in companies], width, label='FY 21-22', color='#2E86AB', alpha=0.8)
bars2_param = ax2.bar(x, [parameters_available[company][1] for company in companies], width, label='FY 22-23', color='#A23B72', alpha=0.8)
bars3_param = ax2.bar(x + width, [parameters_available[company][2] for company in companies], width, label='FY 23-24', color='#F18F01', alpha=0.8)

ax2.set_xlabel('Companies', fontsize=12, fontweight='bold')
ax2.set_ylabel('Number of Parameters Available', fontsize=12, fontweight='bold')
ax2.set_title('ESG Parameters Availability by Company Across Fiscal Years', fontsize=14, fontweight='bold', pad=20)
ax2.set_xticks(x)
ax2.set_xticklabels(companies, rotation=45, ha='right')
ax2.legend(loc='upper left', fontsize=10)
ax2.grid(axis='y', alpha=0.3)
ax2.set_ylim(0, 26)

# Add value labels on bars for parameters
def add_param_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                f'{int(height)}', ha='center', va='bottom', fontsize=8)

add_param_labels(bars1_param)
add_param_labels(bars2_param)
add_param_labels(bars3_param)

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nSummary Statistics:")
print("\n1. Total Scores Analysis:")
for company in companies:
    scores = total_scores[company]
    print(f"{company}: Mean = {np.mean(scores):.3f}, Trend = {scores[2] - scores[0]:.3f}")

print("\n2. Parameter Availability Analysis:")
for company in companies:
    params = parameters_available[company]
    print(f"{company}: Mean = {np.mean(params):.1f}/26, Trend = {params[2] - params[0]:.0f}")
```
